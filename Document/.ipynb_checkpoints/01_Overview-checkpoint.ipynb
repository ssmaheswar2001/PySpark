# Overview
- PySpark is the Python API for Apache spark.
- It enables us to perform real-time, large scale data processing in a distributed environment using python.
- It also provides a PySpark shell for interactively analyzing our data.
- PySpark combiens Python's learnability and ease of use with the power of Apache Spark to enable processing and analysis of data at any size.
- PySpark supports all of Spark's features such as Spark SQL, DataFrames, Structured Streaming, Machine Learning (MLib) and Spark Core.
- 

## Python Spark Connect client
- Spark connect is a client server architecture within Apache Spark that enabes remote connectivity to Spark clusters from any application.
- PySpark provides the client for the Spark Connect server, allowing Spark to be used as a service.

## Spark SQL andd DataFrames
- Spark SQL is Apache Spark's module for working with structured data. It allows us to seamlessly mix SQL queries with Spark programs.
- With PySpark DataFrames we can efficiently read, write, transform, and analyze data using Python and SQL. Whether we use Python or SQL, the same underlying execution engine is used so we will always leverage the full power of Spark.

## Pandas API on Spark
- Pandas API on Spark allows us to scale our pandas workload to any size by running it distributed across multiple nodes.
- We can have a single codebase that works both with pandas (tests, smaller datasets) and with Spark (production, distributed datasets)

## Machine LEarning (MLib)
- Built on top of Spark, MLib is a scalable machine learning library that provides a uniform set of high-level APIs that help users create and tune practical machine learning pipelines.

## Spark Core and RDDs
- Spark Core is the underlying general execution engine for the Spark platform that all other functionality is built on top of.
- It provides RDDs (Resilient Distributed Datasets) and in-memory computing capabilities.
- RDD API is a low-level API which can be difficult to use adn we don not get the benefit of Spark's automatic query optimization capabilites. We recommend using DataFrames (Spark SQL and DataFrames) instead of RDDs as it allows us to express what we want more easily and lets Spark automatically contruct the most efficient query for us.

## Spark Streaming (Legacy)
- Spark Streaming is an extension of the Core Spark API that enables scalable, hihg-throughput, fault tolerance stream processing of live data streams.
- Spark Streaming is the previous version, a newer easier to use streaming engine in Spark called Structured Streaming which we should use for our streaming applications and pipelines.
